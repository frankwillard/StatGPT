```python
import pandas as pd
import numpy as np


```

```python
df = pd.read_csv('590.csv')
df.head()
```

```python
# hall of fame data

df = df[["Player", "PTS_per_game", "TRB_per_game" , "AST_per_game" , "STL_per_game" ,"BLK_per_game", "Hall_of_Fame"]]

df.rename(columns={'PTS_per_game': 'Points', 'TRB_per_game': 'Rebounds', 'AST_per_game': 'Assists', 'STL_per_game': 'Steals', 'BLK_per_game': 'Blocks', "Hall_of_Fame" : "hof"}, inplace=True)


# Convert all columns to integer type
df = df.astype(float, errors='ignore')


cond1 = (df["Points"] < 99) & (df["Rebounds"] < 99) & (df["Assists"] < 99) & (df["Steals"] < 99) & (df["Blocks"] < 99)

cond2 = (df["Points"] > 0) & (df["Rebounds"] > 0) & (df["Assists"] > 0) & (df["Steals"] > 0) & (df["Blocks"] > 0)



df = df[cond1]
df = df[cond2]

df.to_csv('hof.csv', index=False)
```

```python
print(df["Points"].mean(), df["Rebounds"].mean(), df["Assists"].mean(), df["Steals"].mean(), df["Blocks"].mean())
```
